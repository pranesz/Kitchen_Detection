```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.applications import MobileNetV2
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow import keras
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img_height, img_width = 224, 224  # Set according to your needs
batch_size = 32
base_dir = 'images' 
```

```python
# Define your ImageDataGenerator with brightness adjustment
datagen = ImageDataGenerator(
    rotation_range=40,       # Randomly rotate images in the range (degrees)
    width_shift_range=0.2,   # Randomly shift images horizontally (20% of width)
    height_shift_range=0.2,  # Randomly shift images vertically (20% of height)
    shear_range=0.2,         # Shear angle in counter-clockwise direction (in degrees)
    zoom_range=0.2,          # Randomly zoom into images
    horizontal_flip=True,     # Randomly flip images
    fill_mode='nearest',      # Points outside the boundaries are filled according to the nearest pixel
    rescale=1.0/255.0,
    brightness_range=[0.8, 1.2],  # Adjust brightness: [min, max]
    validation_split=0.2
)

# Create your training and validation generators as before
train_generator = datagen.flow_from_directory(
    base_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',  # Binary classification
    subset='training'
)

validation_generator = datagen.flow_from_directory(
    base_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',  # Binary classification
    subset='validation'
)

def display_samples(generator):
    images, labels = next(generator)
    plt.figure(figsize=(10, 10))
    for i in range(4):
        plt.subplot(2, 2, i + 1)
        plt.imshow(images[i])
        label = 'Clean' if labels[i] == 0 else 'Not Clean'
        plt.title(label)
        plt.axis('off')
    plt.show()

# Display samples with brightness adjustment
display_samples(train_generator)
```

```python
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import cv2

# Define a custom function for contrast adjustment
def adjust_contrast(image, factor=1.5):
    # Ensure the image is in float format for processing
    image = image.astype(np.float32)  # Convert to float32 for processing
    # Adjust contrast
    img = cv2.convertScaleAbs(image, alpha=factor, beta=0)
    return img.astype(np.float32)  # Return as float32 to prevent type issues

# Wrap the function to fit the expected input/output format
def contrast_preprocessing_function(x):
    factor = np.random.uniform(0.8, 1.2)  # Random factor for contrast
    return adjust_contrast(x, factor)

# Define your ImageDataGenerator with contrast adjustment
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    rescale=1.0 / 255.0,  # Rescale after contrast adjustmento
    validation_split=0.2,
    preprocessing_function=contrast_preprocessing_function
)

# Create your training and validation generators
train_generator = datagen.flow_from_directory(
    base_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',  # Binary classification
    subset='training'
)

validation_generator = datagen.flow_from_directory(
    base_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',  # Binary classification
    subset='validation'
)

# Display some augmented images to visualize contrast adjustment
def display_samples(generator):
    images, labels = next(generator)
    plt.figure(figsize=(10, 10))
    for i in range(4):
        plt.subplot(2, 2, i + 1)
        plt.imshow(images[i])
        label = 'Clean' if labels[i] == 0 else 'Not Clean'
        plt.title(label)
        plt.axis('off')
    plt.show()

# Display samples with contrast adjustment
display_samples(train_generator)
```

```python
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import cv2

# Function to perform random cropping
def random_crop(image, target_height, target_width):
    height, width, _ = image.shape
    # Generate random coordinates for cropping
    top = np.random.randint(0, height - target_height + 1)
    left = np.random.randint(0, width - target_width + 1)
    # Crop the image
    cropped_image = image[top:top + target_height, left:left + target_width]
    return cropped_image

# Custom preprocessing function that applies random cropping
def random_crop_preprocessing_function(x):
    target_height, target_width = 224, 224  # Specify your target crop size
    return random_crop(x, target_height, target_width)

# Define your ImageDataGenerator with random cropping
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    rescale=1.0 / 255.0,
    validation_split=0.2,
    preprocessing_function=random_crop_preprocessing_function
)

# Create your training and validation generators
train_generator = datagen.flow_from_directory(
    base_dir,
    target_size=(img_height, img_width),  # Set this to the desired final image size after cropping
    batch_size=batch_size,
    class_mode='binary',  # Binary classification
    subset='training'
)

validation_generator = datagen.flow_from_directory(
    base_dir,
    target_size=(img_height, img_width),  # Set this to the desired final image size after cropping
    batch_size=batch_size,
    class_mode='binary',  # Binary classification
    subset='validation'
)

# Display some augmented images to visualize random cropping
def display_samples(generator):
    images, labels = next(generator)
    plt.figure(figsize=(10, 10))
    for i in range(4):
        plt.subplot(2, 2, i + 1)
        plt.imshow(images[i])
        label = 'Clean' if labels[i] == 0 else 'Not Clean'
        plt.title(label)
        plt.axis('off')
    plt.show()

# Display samples with random cropping
display_samples(train_generator)
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import Adam

# Load the MobileNetV2 model
base_model = MobileNetV2(input_shape=(img_height, img_width, 3), include_top=False, weights='imagenet')

# Unfreeze the top layers (you can adjust how many layers to unfreeze)
for layer in base_model.layers[-30:]:  # Unfreeze the last 30 layers
    layer.trainable = True

# Create a new model and add the base model
model = Sequential([
    base_model,
    Flatten(),
    Dense(1, activation='sigmoid')  # Binary classification
])

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(train_generator, validation_data=validation_generator, epochs=10)
```

### Fine-tuning the MobileNetV2 model

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import Adam

# Load the MobileNetV2 model
base_model = MobileNetV2(input_shape=(img_height, img_width, 3), include_top=False, weights='imagenet')

# Freeze all layers initially
for layer in base_model.layers:
    layer.trainable = False
# Create a new model and add the base model
model = Sequential([
    base_model,
    Flatten(),
    Dense(1, activation='sigmoid')  # Binary classification
])
# Compile the model with a higher learning rate for the new layers
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Train the model for a few epochs
model.fit(train_generator, validation_data=validation_generator, epochs=5)

# Unfreeze the last few layers of the base model
for layer in base_model.layers[-30:]:  # Adjust the number of layers to unfreeze
    layer.trainable = True

# Compile the model again with a smaller learning rate
model.compile(optimizer=Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['accuracy'])

# Train the model again with fine-tuning
model.fit(train_generator, validation_data=validation_generator, epochs=10)
```

```python
# Evaluate the model on validation/test data
val_loss, val_accuracy = model.evaluate(validation_generator)
print(f"Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_accuracy:.4f}")
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', patience=3)
model.fit(train_generator, validation_data=validation_generator, epochs=10, callbacks=[early_stopping])
```
